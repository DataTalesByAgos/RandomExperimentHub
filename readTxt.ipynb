```python
from google.colab import drive
drive.mount('/content/drive')
import os

available_files = [f for f in os.listdir('/content/drive/My Drive/') if f.endswith('.txt')]

print("Available txt files:")
for i, file in enumerate(available_files):
    print(f"{i+1}. {file}")

option = int(input("Choose the number of the file you want to read: ")) - 1

if option < 0 or option >= len(available_files):
    print("Invalid option.")
else:
    chosen_file = available_files[option]

    with open(os.path.join('/content/drive/My Drive/', chosen_file), 'r') as f:
        contents = f.read()

    print(contents)
```
