```python
from google.colab import drive
from io import open

drive.mount('/content/drive')

filename = input("Enter the file name (without extension): ") + ".txt"
filepath = f'/content/drive/My Drive/{filename}'

# Create the text to write into the file
newText = f'different text.. this is the {filename} test version'
# Get the file name without the extension
file_name = filename.split('.')[0]

# Open+Write
with open(filepath, 'w') as f:
    f.write(f"File name: {file_name}\n")
    f.write(newText)

# Verify
import os
print("File created:", os.path.exists(filepath))
```
